# Conclusión gráfico 2
Luego de analizar los datos obtenidos de el mes de noviembre de 2017, se identifican los 10 barrios con mayor promedio de finalización de viajes, indicando una alta demanda hacia esas zonas. El barrio con mayor promedio de viajes fue Loop, seguido de River North, Streeterville, West Loop y O'Hare, siendo este el top 5 de los barrios con mayor finalización de viajes. Estos datos dan pie a analizar el por qué de estos resultados, ya que entre los primeros cuatro barrios abarcan la mayoría de promedio de viajes, como por ejemplo que oferta tienen estos barrios, si cultural o gastronómica para influir en la predilección. 

Adicional, se podría hacer un estudio sobre los barrios donde se tiene un mayor inicio de viajes, para poder movilizar mayor cantidad de taxis a esas zonas y así asegurar servicios.

```python
df7.info()
df7.head(10)
```

```python
"""
Prueba la hipótesis:

La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.

Decide por tu cuenta dónde establecer el nivel de significación (alfa). 
"""

# preparación de los datos: sábados con buen clima vs mal clima
good_weather = df7[df7['weather_conditions']=='Good']['duration_seconds']
bad_weather = df7[df7['weather_conditions']=='Bad']['duration_seconds']

# Prueba t de muestras independientes
t_stat, p_value =stats.ttest_ind(good_weather, bad_weather)

alpha = 0.05 # se define el alfa como el 5%

if p_value < alpha:
  print('El p valor es', p_value)
  print('Se rechaza la hipotesis nula. Existe diferencia en la duracion promedio de los viajes segun las condiciones climaticas en los dias sabados')
else:
  print('No se rechaza la hipotesis nula. No se encuentra diferencia significativa en la duracion promedio de los viajes en los dias sabados')
```

# Conclusión prueba de hipótesis
Inicialmente, se define un nivel de significacia del 5% (alfa=0.05). Con los datos previamente obtenidos de la duración promedio de los 
viajes desde el barrio Loop (el que tuvo mayor promedio de finalización de viajes) hasta el aeropuerto internacional O'Hare se evaluó la hipótesis nula que la duración de viajes promedio no se ve afectada por las condiciones climáticas como la lluvia. Para esto se realizó una prueba t y se rechazó dicha hipótesis, viendo que si hay diferencias significativas entre la duración promedio de viajes debido al clima, algo que tiene mucha lógica ya que se vuelve más riesgosa la conducción (como un frenado poco eficiente o baja visibilidad), por lo que los taxistas optan por disminuir la velocidad y así prevenir accidentes.
